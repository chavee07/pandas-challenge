```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
# Total number of schools by unique value, len returns count for unique values...
# .. w/o len, will return a ls[ of numbers, of entire, count(15),etc. starting at 0]
total_schools=len(school_data_complete['School ID'].unique())
print(f"Total number of unique schools in District:", total_schools)

total_studentid=len(school_data_complete['Student ID'].unique())
print(f"Total number of students in District:",total_studentid)

total_budget=school_data['budget'].sum()
print(f"Total budget for District:", total_budget)

average_mathscore= school_data_complete['math_score'].mean()
print(f"District Average math score:", average_mathscore)

average_readingscore=school_data_complete['reading_score'].mean()
print(f"District Average reading score:", average_readingscore )

print('-------------------------------------------------------------- ')
#not using averages, just using math score column to calc., find all math scores over 70.
# first line will return math scores in a new column with math scores >=70 and index.
# need to know a count(number), here the DF is not useful alone, 
# so I use it to create a count of passing grades for later use in percent calc.
passing_math=student_data.loc[student_data['math_score']>=70,['math_score']]
# passing_math

# I can then use this dataframe to count the value of math scores that are over >=70.
# will print a number I can use to solve(and plug into calc. to find) percentage
pass_math_count= passing_math['math_score'].count()

print(f"Total amount of students with passing math grades in District:", pass_math_count)

# divide the count of passing math grades over the total students 
percent_passingmath=pass_math_count/total_studentid*100
percent_passingmath
print(f"Total percent of students with passing math grades in District:", percent_passingmath)


print('-------------------------------------------------------------- ')

# Calc. for Reading scores

passing_reading=student_data.loc[student_data['reading_score']>=70,['reading_score']]
# passing_reading

pass_reading_count= passing_reading['reading_score'].count()
# pass_reading_count
print(f"Total amount of students with passing reading grades in District:", pass_reading_count)

percent_passingreading=pass_reading_count/total_studentid*100
percent_passingreading
print(f"Total percent of students with passing math grades in District:", percent_passingreading)

print('-------------------------------------------------------------- ')

overall_passing=(percent_passingmath+percent_passingreading)/2
print(f"The overall passing rate for the District is:" , overall_passing)

summary_df=pd.DataFrame({"Total Schools": [total_schools],
                         "Total Students": [total_studentid],
                         "Budget ": [total_budget], 
                         "Average Math Score ": [average_mathscore],
                         "Average Reading Score": [average_readingscore],
                         "% Passing Math": [percent_passingmath],
                         "% Passing Reading": [percent_passingreading],
                         "% Passing Overall": [overall_passing]
                               })
summary_df

# formatting required
districtsummary_df=pd.DataFrame(summary_df, columns=["Total Schools",
                                           "Total Students",
                                           "Budget ",
                                           "Average Math Score ",
                                           "Average Reading Score",
                                           "% Passing Math","% Passing Reading",
                                           "% Passing Overall"])
districtsummary_df['Total Students']= districtsummary_df['Total Students'].map('{:,}'.format)
# districtsummary_df['Budget']= districtsummary_df['Budget'].map('${:,.2f}'.format) 



districtsummary_df["% Passing Math"] = districtsummary_df["% Passing Math"].map("{:,.2f}%".format)
districtsummary_df["% Passing Reading"] = districtsummary_df["% Passing Reading"].map("{:,.2f}%".format)
districtsummary_df["% Passing Overall"] = districtsummary_df["% Passing Overall"].map("{:,.2f}%".format)

districtsummary_df
```

```python
# avverage scores of all high schools and grades
average_scores_by_highschool=student_data.groupby(['school_name', 'grade']).mean().reset_index()
# average_scores_by_highschool
```

```python
# use loc to isolate individual grades in reading and math scores(includes student id).
ninth_grade_df = average_scores_by_highschool.loc[average_scores_by_highschool['grade'] == '9th']
# ninth_grade_df

tenth_grade_df=average_scores_by_highschool.loc[average_scores_by_highschool['grade']=='10th']
# tenth_grade_df

eleventh_grade_df = average_scores_by_highschool.loc[average_scores_by_highschool['grade'] == '11th']
# eleventh_grade_df

twelfth_grade_df = average_scores_by_highschool.loc[average_scores_by_highschool['grade'] == '12th']
# twelfth_grade_df
```

```python
# reduce dataframes to school, math score, and reading score per grade.

ninth_grade_red_df = pd.DataFrame({"school": ninth_grade_df["school_name"],
                                   "reading_score_9th": ninth_grade_df['reading_score'],
                                   "math_score_9th": ninth_grade_df['math_score']})
ninth_grade_red_df

tenth_grade_red_df = pd.DataFrame({"school": tenth_grade_df["school_name"],
                                   "reading_score_10th": tenth_grade_df['reading_score'],
                                   "math_score_10th": tenth_grade_df['math_score']})
tenth_grade_red_df

eleventh_grade_red_df = pd.DataFrame({"school": eleventh_grade_df["school_name"],
                                   "reading_score_11th": eleventh_grade_df['reading_score'],
                                   "math_score_11th": eleventh_grade_df['math_score']})
eleventh_grade_red_df

twelfth_grade_red_df = pd.DataFrame({"school": twelfth_grade_df["school_name"],
                                   "reading_score_12th": twelfth_grade_df['reading_score'],
                                   "math_score_12th": twelfth_grade_df['math_score']})
# twelfth_grade_red_df
```

```python
all_scores_by_grade_hs_df = pd.merge(ninth_grade_red_df, tenth_grade_red_df, on = 'school')\
.merge(eleventh_grade_red_df, on = 'school')\
.merge(twelfth_grade_red_df, on = 'school') 
all_scores_by_grade_hs_df
```
